```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Generate a random signal
N = 1024  # Signal length
x = np.random.rand(N)

# Define the DFT function
def dft(x):
    N = len(x)
    X = np.zeros(N, dtype=complex)
    for k in range(N):
        for n in range(N):
            X[k] += x[n] * np.exp(-2j * np.pi * k * n / N)
    return X

# Measure DFT computation time
start_time = time.time()
X_dft = dft(x)
dft_time = time.time() - start_time
print(f"DFT computation time: {dft_time:.6f} seconds")

# Measure FFT computation time
start_time = time.time()
X_fft = np.fft.fft(x)
fft_time = time.time() - start_time
print(f"FFT computation time: {fft_time:.6f} seconds")

# Compare results
print(f"FFT is {dft_time / fft_time:.2f} times faster than DFT")

# Plot the random signal
plt.figure(figsize=(12, 6))

# Plot the signal
plt.subplot(1, 2, 1)
plt.plot(x)
plt.title('Random Signal')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Plot the frequency spectrum
plt.subplot(1, 2, 2)
frequencies = np.fft.fftfreq(N)
plt.plot(frequencies, np.abs(X_fft))
plt.title('Frequency Spectrum (FFT)')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.xlim(0, 0.5)  # Limit x-axis for better visualization
plt.grid()

plt.tight_layout()
plt.show()
```
