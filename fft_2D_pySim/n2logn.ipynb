```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for N
N = np.arange(1, 1025, 1)

# Compute N^2, logN, and N^2*logN
N_squared = N**2
logN = np.log2(N)
N_squared_logN = N**2 * logN

# Plot the results
plt.figure(figsize=(8, 5))
plt.plot(N, N_squared, label='$N^2$', color='b')
plt.plot(N, logN, label='$\log N$', color='g')
plt.plot(N, N_squared_logN, label='$N^2 \log N$', color='r')
plt.xlabel('N', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.title('Comparison of $N^2$, $\log N$, and $N^2 \log N$', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```
